```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/ventes-data.csv")
```

```python
df
```

```python
v = df.groupby("Month")[["Revenue"]].sum().sort_values(["Revenue"])
v
```

```python
df.info()
```

```python
top = df.groupby(["Month","Product"])["Units Sold"].sum()
top5 = top.groupby("Month", group_keys = False).nlargest(5)
top5
```

```python
# Method 2
tope = df.groupby(["Month","Product"])[["Units Sold"]].sum()
topI = tope.sort_values(by = ["Month", "Units Sold"], ascending = [True, False])
tops = topI.groupby("Month").head(5)
tops
```

```python
bas = v.head(3)
eleve = v.tail(3)
print(f"Les mois avec les ventes les plus eleves {eleve} \n les mois avec les ventes les plus basses {bas}")
```
